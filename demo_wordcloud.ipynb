```python
Platform_list = topbest['Platform'].unique()
na_rev, eu_rev, jp_rev, ot_rev, gb_rev = [], [], [], [], []

for pl in Platform_list:
    na_rev.append(topbest[topbest['Platform'] == pl]['NA_Sales'].sum())
    eu_rev.append(topbest[topbest['Platform'] == pl]['EU_Sales'].sum())
    jp_rev.append(topbest[topbest['Platform'] == pl]['JP_Sales'].sum())
    ot_rev.append(topbest[topbest['Platform'] == pl]['Other_Sales'].sum())
    gb_rev.append(topbest[topbest['Platform'] == pl]['Global_Sales'].sum())
    
plventes = pd.DataFrame({
    'Platform': Platform_list,
    'na': na_rev,
    'eu': eu_rev,
    'jp': jp_rev,
    'other': ot_rev,
    'global': gb_rev
})

plventes.head()
```

```python
plt.style.use("fivethirtyeight")
sns.barplot(data=plventes[:10], x='Platform', y='na', palette='bone')
plt.xticks(fontsize=12)
plt.xlabel("Platforme")
plt.ylabel("Jeux vendus")
plt.title("Top 10 Platformes Amerique du nord")
plt.show();

plt.style.use("fivethirtyeight")
sns.barplot(data=plventes[:10], x='Platform', y='jp', palette='mako')
plt.xticks(fontsize=12)
plt.xlabel("Platforme")
plt.ylabel("Jeux Vendus")
plt.title("Top-10 plateformes japon")
plt.show();

plt.style.use("fivethirtyeight")
sns.barplot(data=plventes[:10], x='Platform', y='eu', palette='viridis')
plt.xticks(fontsize=12)
plt.xlabel("Platforme")
plt.ylabel("Jeux vendus")
plt.title("Top-10 Platformes Europe")
plt.show();
```

```python
genre_list = topbest['Genre'].unique()
na_rev, eu_rev, jp_rev, ot_rev, gb_rev = [], [], [], [], []

for gen in genre_list:
    na_rev.append(topbest[topbest['Genre'] == gen]['NA_Sales'].sum())
    eu_rev.append(topbest[topbest['Genre'] == gen]['EU_Sales'].sum())
    jp_rev.append(topbest[topbest['Genre'] == gen]['JP_Sales'].sum())
    ot_rev.append(topbest[topbest['Genre'] == gen]['Other_Sales'].sum())
    gb_rev.append(topbest[topbest['Genre'] == gen]['Global_Sales'].sum())
    
gventes = pd.DataFrame({
    'genre': genre_list,
    'na': na_rev,
    'eu': eu_rev,
    'jp': jp_rev,
    'other': ot_rev,
    'global': gb_rev
})

gventes.head()
```

```python
plt.style.use("ggplot")
sns.barplot(data=gventes[:10], x='genre', y='na', palette='Accent_r')
plt.xticks(rotation=90, fontsize=10)
plt.xlabel("Genre")
plt.ylabel("Number of North American Sales")
plt.title("Top-10 Genres N.Américain")
plt.show()

plt.style.use("ggplot")
sns.barplot(data=gventes[:10], x='genre', y='jp', palette='mako')
plt.xticks(rotation=90, fontsize=10)
plt.xlabel("Genre")
plt.ylabel("JP Sales")
plt.title("Top-10 Genres japonais")
plt.show()

plt.style.use("ggplot")
sns.barplot(data=gventes[:10], x='genre', y='eu', palette='mako')
plt.xticks(rotation=90, fontsize=10)
plt.xlabel("Genre")
plt.ylabel("EU Sales")
plt.title("Top-10 Genre Europe")
plt.show()
```
